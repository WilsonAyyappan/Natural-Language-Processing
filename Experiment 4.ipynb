### Importing Libraries:

We have to import specific python libraries that provide necessary functions and tools for handling the data and performing analyses.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Dataset:

I have chosen to use Hugging Face Datasets' "surrey-nlp/PLOD-CW" dataset for this investigation which is available in Surrey learn.

```python
# Loading the "surrey-nlp/PLOD-CW" dataset from Hugging Face.
from datasets import load_dataset, load_metric
dataset = load_dataset("surrey-nlp/PLOD-CW")

# Convert the 'train' split of the PLOD-CW dataset to a Pandas DataFrame for easier manipulation and analysis.
df = dataset['train'].to_pandas()
```

### Data Preparation and Performance Comparison:

- **Data Preparation:**
Text data from the dataset is vectorized using TF-IDF, and labels are encoded and transformed into one-hot encoding.
- **Model Construction:**
Two separate 1D CNN models are built, each using a different loss function: Categorical Crossentropy and Sparse Categorical Crossentropy.
- **Training:**
Both models are trained on the TF-IDF vectorized data with identical hyperparameters and training procedures.
- **Evaluation:**
The performance of each model is evaluated using standard metrics such as accuracy, precision, recall, and F1-score on a separate test dataset.
- **Visualization:**
The training history (loss and accuracy) of each model is plotted to observe the training process and convergence behavior.
- **Analysis:**
Classification reports and confusion matrices are generated to analyze the performance of each model and compare their effectiveness in classifying text data.

```python
# Combine tokens into single strings per sample
train_texts = [" ".join(tokens) for tokens in dataset['train']['tokens']]
test_texts = [" ".join(tokens) for tokens in dataset['test']['tokens']]

# Vectorize the data
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  
X_train_tfidf = tfidf_vectorizer.fit_transform(train_texts).toarray()
X_test_tfidf = tfidf_vectorizer.transform(test_texts).toarray()

# Encode labels 
all_possible_tags = [tag for sample_tags in dataset['train']['ner_tags'] + dataset['test']['ner_tags'] for tag in sample_tags]
label_encoder = LabelEncoder()
label_encoder.fit(all_possible_tags)
y_train = label_encoder.transform([tags[1] for tags in dataset['train']['ner_tags']])  
y_test = label_encoder.transform([tags[1] for tags in dataset['test']['ner_tags']])

# Convert labels to one-hot encoding
y_train_encoded = to_categorical(y_train, num_classes=len(label_encoder.classes_))
y_test_encoded = to_categorical(y_test, num_classes=len(label_encoder.classes_))

# Reshape TF-IDF vectors to include an additional dimension for timesteps
X_train_tfidf_reshaped = np.expand_dims(X_train_tfidf, axis=2)  # Add channel dimension for 1D CNN
X_test_tfidf_reshaped = np.expand_dims(X_test_tfidf, axis=2)

# Define 1D CNN model builder function
def build_cnn_model(input_shape, num_classes, loss_function='categorical_crossentropy'):
    model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape),
        GlobalMaxPooling1D(),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss=loss_function, metrics=['accuracy'])
    return model

input_shape = (X_train_tfidf_reshaped.shape[1], X_train_tfidf_reshaped.shape[2])
num_classes = len(label_encoder.classes_)

# Model with categorical_crossentropy
model_cat = build_cnn_model(input_shape, num_classes, 'categorical_crossentropy')
history_cat = model_cat.fit(X_train_tfidf_reshaped, y_train_encoded, validation_split=0.2, epochs=10, batch_size=32)

# Model with sparse_categorical_crossentropy
model_sparse = build_cnn_model(input_shape, num_classes, 'sparse_categorical_crossentropy')
history_sparse = model_sparse.fit(X_train_tfidf_reshaped, y_train, validation_split=0.2, epochs=10, batch_size=32)

# Function to plot history
def plot_history(history, title_prefix=''):
    epochs = range(1, len(history.history['accuracy']) + 1)
    
    plt.figure(figsize=(12, 5))
    
    # Plot training and validation accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history.history['accuracy'], 'bo', label='Training acc')
    plt.plot(epochs, history.history['val_accuracy'], 'b', label='Validation acc')
    plt.title(f'{title_prefix} Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Plot training and validation loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history.history['loss'], 'ro', label='Training loss')
    plt.plot(epochs, history.history['val_loss'], 'r', label='Validation loss')
    plt.title(f'{title_prefix} Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Plot history for both models
plot_history(history_cat, 'Categorical Crossentropy')
plot_history(history_sparse, 'Sparse Categorical Crossentropy')

# Function to plot confusion matrix
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(title)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.show()

# Evaluate categorical_crossentropy model
y_pred_cat = np.argmax(model_cat.predict(X_test_tfidf_reshaped), axis=1)
print("Classification Report for categorical_crossentropy:")
print(classification_report(y_test, y_pred_cat)) 
cm_cat = confusion_matrix(y_test, y_pred_cat)
plot_confusion_matrix(cm_cat, "Confusion Matrix for categorical_crossentropy")

# Evaluate sparse_categorical_crossentropy model
y_pred_sparse = model_sparse.predict(X_test_tfidf_reshaped)
y_pred_sparse_labels = np.argmax(y_pred_sparse, axis=1)  # Get class labels
print("Classification Report for sparse_categorical_crossentropy:")
print(classification_report(y_test, y_pred_sparse_labels))
cm_sparse = confusion_matrix(y_test, y_pred_sparse_labels)
plot_confusion_matrix(cm_sparse, "Confusion Matrix for sparse_categorical_crossentropy")
```

### Conclusion:

- **Accuracy:**
Both models achieved an accuracy of 88%, indicating that they correctly classified the majority of instances in the test dataset.
Precision, Recall, and F1-Score: For class 2, which has the highest support (number of instances), both models achieved high precision, recall, and F1-score, indicating robust performance in correctly classifying this class. However, for classes 0, 1, and 3, the precision, recall, and F1-scores are notably lower, indicating challenges in correctly classifying instances belonging to these classes.
- **Imbalance in Classes:**
Classes 0, 1, and 3 have very low support (few instances), which results in low precision, recall, and F1-scores for these classes. This imbalance may affect the overall performance metrics of the models.
- **Effectiveness of Loss Functions:**
Interestingly, despite the difference in loss functions used (Categorical Crossentropy vs. Sparse Categorical Crossentropy), both models exhibit similar performance across all metrics. This suggests that, in this particular classification task and dataset, the choice of loss function may not significantly impact model performance.
